```python
import pandas as pd
import numpy as np
import json 
import copy
```

```python
# Importing the olympics event detials, athlete-wise
ath_events = pd.read_csv('athlete_events.csv')
ath_events.head()


e = ath_events[ath_events['Medal'].isin(['Gold'])]
e[e['NOC'].isin(['IND'])]
```

```python
# Importing the country NOC codes that olympics uses as a standard
con_codes = pd.read_csv('country_codes.csv')
con_codes.head()
```

```python
# Importing in the latitude and longitude data of all the countries of the world and making a nested list 
f = open('country_lat_long.JSON')
# NOTE: The original file was edited by me (Sid) to make it compatitable with the non-official names used for 
# countries by the olympics (Eg: Syria, instead of "Syrian Arab Republic")
con_lat_long = json.load(f)

info = []
for i in con_lat_long['ref_country_codes']:
    info.append([i['country'], i['latitude'], i['longitude']])
info
```

```python
# Converting the nested list into a Dictionary
country_info = dict()
for i in info:
    country_info[i[0]] = i[1:]
country_info
back = copy.deepcopy(country_info)
back
```

```python
# Adding the NOC codes to their respective countries
for index, row in con_codes.iterrows():
    try:
        data = back[row['Country']]
        data.append(row['NOC'])
        back[row['Country']] = data
        
    except:
        back[row['Country']] = [-1,-1, row['NOC']]

back
```

```python
con_info = copy.deepcopy(back)
con_info    # HAS: Country name mapped to LAT, LONG, NOC
```

```python
gmeds = copy.deepcopy(con_info)
# Adding gold medals count to the dictionary, init to 0
for i in gmeds.keys():
    data = gmeds[i]
    data.append(0)
    gmeds[i] = data
    
gmeds
```

```python
# changing the keys of the dictionary from country to NOC code instead (to use with ath_Events)
gmeds_NOC = dict()
for i in gmeds.keys():
    data = gmeds[i]
    if (type(data[-2]) != str):
        continue
        
    gmeds_NOC[data[-2]] = [i, data[0], data[1], 0]
    
    
gmeds_NOC
```

```python
# Counting and adding up all the gold medals that countries have achieved
for index, row in ath_events[ath_events['Medal'].isin(['Gold'])].iterrows():        
    if (row['NOC'] not in gmeds_NOC.keys()):
        continue
    
    data = gmeds_NOC[row['NOC']]
    data[-1] += 1
    gmeds_NOC[row['NOC']] = data

gmeds_NOC
```

```python
# Adding a silver medal count to the dictionary
gs_meds_NOC = copy.deepcopy(gmeds_NOC)
for i in gs_meds_NOC.keys():
    data = gs_meds_NOC[i]
    data.append(0)
    
gs_meds_NOC
```

```python
# Counting and adding up all the silver medals that countries have achieved
for index, row in ath_events[ath_events['Medal'].isin(['Silver'])].iterrows():
    if (row['NOC'] not in gs_meds_NOC.keys()):
        continue;
    
    data = gs_meds_NOC[row['NOC']]
    data[-1] += 1
    gs_meds_NOC[row['NOC']] = data

gs_meds_NOC
```

```python
# Adding a bronze medal count to the dictionary
gsb_meds_NOC = copy.deepcopy(gs_meds_NOC)
for i in gsb_meds_NOC.keys():
    data = gsb_meds_NOC[i]
    data.append(0)

gsb_meds_NOC
```

```python
# Counting and adding up all the bronze medals that countries have achieved
for index, row in ath_events[ath_events['Medal'].isin(['Bronze'])].iterrows():
    if (row['NOC'] not in gsb_meds_NOC.keys()):
        continue;
    
    data = gsb_meds_NOC[row['NOC']]
    data[-1] += 1
    gsb_meds_NOC[row['NOC']] = data

gsb_meds_NOC
```

```python
gsb_meds_NOC
```

```python
# Finally, we convert the dictionary into a JSON file by building a string with a loop. The schema was prepared
# such that it would work with the map API we used.
js = '''{
    "data" : [
    
'''
s=''
for k in gsb_meds_NOC.keys():
    data = gsb_meds_NOC[k]
    if (not data[-1] and not data[-2] and not data[-3]):
        continue
    
    js += '{' + '\n'
    js += '\t  ' + '"Country_name":' + '"' + data[0] + '"' + ',' + '\n'
    js += '\t  ' + '"NOC":' + '"' + k  + '"' + ',' + '\n'
    js += '\t  ' + '"lat":' + str(data[1])  + ',' + '\n'
    js += '\t  ' + '"long":' + str(data[2])  + ',' + '\n'
    js += '\t  ' + '"Gold_Medals":' + str(data[3])  + ',' + '\n'
    js += '\t  ' + '"Silver_Medals":' + str(data[4])  + ',' + '\n'
    js += '\t  ' + '"Bronze_Medals":' + str(data[5]) + '\n'
    js += '\t  ' + '},\n'

js += '  ]' + '\n' + '}'
print(js)
```

```python
# Writing the file
with open('medals_country_latlong.json', 'w') as f:
    f.write(js)
```
